```python
from datasets import load_dataset

ds = load_dataset("statworx/haiku")
```

```python
ds.keys()
```

```python
train_haik = ds['train']
```

```python
import pandas as pd

df = train_haik.to_pandas()
```

```python
df.to_csv('haikus.csv', index=False)
```

```python
ds.save_to_disk('haiku_dataset')
```

```python
# convert all splits to pandas

for split_name in ds.keys():
    df2 = ds[split_name].to_pandas()

df2
```

```python
# get and save as text

with open('haikus_full.txt', 'w', encoding='utf-8') as f:
    for idx, row in df2.iterrows():
        f.write(f"Haiku #{idx + 1}\n")
        f.write(f"Source: {row['source']}\n")
        f.write(f"Text: {row['text']}\n")
        f.write(f"Keywords: {row['keywords']}\n")
        f.write(f"Gruen Score: {row['gruen_score']}\n")
        f.write("\n" + "="*60 + "\n\n")
```
